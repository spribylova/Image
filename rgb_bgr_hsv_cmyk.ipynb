```python
# OpenCV reads in images in BGR format (instead of RGB) because when OpenCV was first being developed,
# BGR color format was popular among camera manufacturers and image software providers.
# The red channel was considered one of the least important color channels, so was listed last,
# and many bitmaps use BGR format for image storage.
# However, now the standard has changed and most image software and cameras use RGB format,
# which is why, in programs, it's good practice to initially convert BGR images to RGB before
# analyzing or manipulating any images.

# In 1987, Microsoft Windows ran on the IBM PS/2, and an early IBM video display controller, VGA,
# made use of the INMOS 171/176 RAMDAC chip, which was easier to use when images were stored in BGR format.
```

```python
# Image is loaded in RGB (Red Green Blue) color space, size 6 kb
# Blue is the least important channel
import cv2
from google.colab.patches import cv2_imshow
rgb = cv2.imread("rgb.jpeg")
rgb=cv2_imshow(rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Converting RGB to BGR, size 6 kb to size 12 kb
# Red is the least importat channel.
# The layout of the Red, Green, and Blue subpixels is the primary distinction between RGB and BGR.
# There is no difference in the vibrancy and accuracy of the colours.
# Converting a RGB image to BGR and vice versa can have several reasons:
# several image processing libraries have different pixel orderings, lower storage size, ...
rgb = cv2.imread("rgb.jpeg")
bgr = cv2.cvtColor(rgb, cv2.COLOR_RGB2BGR)
cv2_imshow(bgr)
cv2.imwrite('bgr.jpeg', bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Converting BGR to HSV (Hue, Saturation, Value), size 6 kb to size 15 kb
hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
cv2.imwrite('hsv.jpeg', hsv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Converting BGR to CMYK, size 6 kb to size 11 kb
import numpy as np
# Make float and divide by 255 to give BGRdash
bgrdash = bgr.astype(np.float64)/255.
# Calculate black K for each color
k = 1 - np.max(bgrdash, axis=2)
c = (1-bgrdash[...,2] - k)/(1-k)
m = (1-bgrdash[...,1] - k)/(1-k)
y = (1-bgrdash[...,0] - k)/(1-k)
# Combine 4 channels into single image and re-scale back up to uint8
cmyk = (np.dstack((c,m,y,k)) * 255).astype(np.uint8)
cv2_imshow(cmyk)
cv2.imwrite('cmyk.jpeg',cmyk )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Converting BGR to gray, size 6 kb to size 10 kb
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
cv2.imwrite('gray.jpeg',gray )
cv2.waitKey(0)
cv2.destroyAllWindows()
```
